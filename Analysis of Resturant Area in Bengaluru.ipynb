####            Remove Those Columns which are not relevent to my Analysis like URL and others.

##### Here we can see that in Rates Columns there some missing Value and also in Location,rest_type,Cuisines as total rows are 51717

# Dropping Dublicate

```python
df.drop_duplicates(inplace=True)
df.shape
```

#### Remove those Rows which are dublicate 

# Clean the Data

```python
df['rate'].unique()
```

# Remove 'New' and '-' from Rate Columns

```python
def remove(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
df['rate']=df['rate'].apply(remove)
df['rate'].head()
```

##### Number of Null value now present in Rate Columns

```python
df.rate.isnull().sum()
```

# Filling the Null Value in Rate Column with Mean

```python
df['rate'].fillna(df['rate'].mean(),inplace=True)
df['rate'].isnull().sum()
```

```python
df.info()
```

# Dropping Null Value 

```python
df.dropna(inplace=True)
df.head()
```

# Rename Columns 

```python
df.rename(columns={'approx_cost(for two people)':'Cost2plates','listed_in(type)':'Type'},inplace=True)
df.head()
```

# Drop listed_in(city) as it is same as location

```python
df=df.drop(['listed_in(city)'],axis=1)
```

```python
df
```

```python
df['Cost2plates'].unique()
```

# Removing (,) from Cost2plates Columns

```python
def remove(value):
    value=str(value)
    if ',' in value:
        value=value.replace(',','')
        return float(value)
    else:
        return float(value)
```

```python
df['Cost2plates']=df['Cost2plates'].apply(remove)
df['Cost2plates'].unique()
```

# Clean Rest_type

```python
df['rest_type'].value_counts()
```

```python
rest_types=df['rest_type'].value_counts(ascending=False)
rest_types
```

# Less than 1000 in Other Columns name rest_types_lessthan1000

```python
rest_types_lessthan1000=rest_types[rest_types<1000]
rest_types_lessthan1000
```

# Making rest Type less than 1000 in frequency as other

```python
def make(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
```

```python
df['rest_type']=df['rest_type'].apply(make)
df['rest_type'].value_counts
```

# cleaning Location Column

```python
df['location'].value_counts()
```

```python
location=df['location'].value_counts(ascending=False)
location_lessthan300=location[location<300]
```

```python
def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
```

```python
df['location']=df['location'].apply(handle_location)
df['location'].value_counts
```

# Cleaning cuisines

```python
cuisines=df['cuisines'].value_counts(ascending=False)
cuisines_lessthan100=cuisines[cuisines<100]
```

```python
def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value
```

```python
df['cuisines']=df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```
